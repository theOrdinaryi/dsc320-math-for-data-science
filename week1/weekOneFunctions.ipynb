```python
import pandas as pd  

housing = pd.read_csv("housing_data.csv")
housing.head()
```

```python
import numpy as np  

def rmse(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    
    squared_errors = (predicted - actual) ** 2
    mse = np.mean(squared_errors)
    return np.sqrt(mse)
```

```python
actual = housing["sale_price"]
predicted = housing["sale_price_pred"]

rmse_value = rmse(actual, predicted)
rmse_value
```

```python
def mae(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    
    abs_errors = np.abs(predicted - actual)
    return np.mean(abs_errors)
```

```python
mae_value = mae(actual, predicted)
mae_value
```

```python
mushrooms = pd.read_csv("mushroom_data.csv")
mushrooms.head()
```

```python
def accuracy(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    
    correct = np.sum(actual == predicted)
    return correct / len(actual)
```

```python
acc_value = accuracy(mushrooms["actual"], mushrooms["predicted"])
acc_value, acc_value * 100
```

```python
def f(p):
    return (0.005 * p**6
            - 0.27 * p**5
            + 5.998 * p**4
            - 69.919 * p**3
            + 449.17 * p**2
            - 1499.7 * p
            + 2028)
```

```python
import matplotlib.pyplot as plt  # already imported, but fine to repeat

p_values = np.linspace(0, 10, 1000)
f_values = f(p_values)

plt.figure(figsize=(8, 5))
plt.plot(p_values, f_values)
plt.xlabel("p")
plt.ylabel("f(p)")
plt.title("Error function f(p)")
plt.grid(True)
plt.show()
```
